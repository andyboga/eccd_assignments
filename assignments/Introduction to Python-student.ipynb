```python
def remove_vowels(s: str) -> str:
    """
    This function removes all vowels from strings
    Example: "hello" -> "hll"
    """

  # Defino las letras que son vocales, incluyendo las que llevan tilde
    vowels = "aeiouáéíóú"

    # Utilizo list comprehension para seleccionar solo los caracteres que no son vocales
    result = [char for char in s if char.lower() not in vowels]

    # Devuelve el string sin las vocales
    return ''.join(result)
```

```python
assert remove_vowels("hola") == "hl"
answer_1  = remove_vowels("este fue un buen día")
print(answer_1)
```

```python
def sum_list_of_floats(a: list, b:list) -> list:
    """
    Returns the coordinate wise sum of a and b.
    """

    # Chequear que las listas tienen la misma longitud
    if len(a) != len(b):
        raise ValueError("Las listas deben tener la misma longitud")

    # Sumar las listas elemento a elemento
    result = [x + y for x, y in zip(a, b)]

    return result
```

```python
assert np.allclose(
    sum_list_of_floats(
        [1, 2, 3], [4, 5, -2.3]),
    [5, 7, 0.7]
)
answer_2 = sum_list_of_floats([0, 0], [1, -1])
print(answer_2)
```
